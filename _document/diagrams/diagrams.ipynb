# Custom

## svg2png

```python
import os
import cairosvg

directory = './icon/svg'
directory_dist = './icon'
files = os.listdir(directory)

for file in files:
    if file.endswith('.svg'):
        svg_path = os.path.join(directory, file)
        png_path = os.path.join(directory_dist, file.replace('.svg', '.png'))
        cairosvg.svg2png(url=svg_path, write_to=png_path)
```
